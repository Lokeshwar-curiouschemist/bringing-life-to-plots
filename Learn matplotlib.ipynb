<h3 style = color:#060270;>Example(1)</h3>

````python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define domain
x = np.linspace(-4 * np.pi, 4 * np.pi, 800)

# 2. Define function(s)
y = 2 * np.sin(2 * x)
# Optional: derivative or second function
# dy = 4 * np.cos(2 * x)

# 3. Set up plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='y = 2¬∑sin(2x)', color='green', linewidth=2)
# Optional: add derivative
# plt.plot(x, dy, '--', label="f'(x) = 4¬∑cos(2x)", color='orange')

# 4. Origin-centered coordinate axes
ax = plt.gca()
ax.spines['bottom'].set_position('zero')   # x-axis at y=0
ax.spines['left'].set_position('zero')     # y-axis at x=0
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

# 5. Optional: axis arrows
ax.plot(1, 0, ">r", transform=ax.get_yaxis_transform(), clip_on=False)  # x-arrow
ax.plot(0, 1, "^r", transform=ax.get_xaxis_transform(), clip_on=False)  # y-arrow

# 6. Labels (repositioned & resized)
ax.set_xlabel("x", fontsize=9, labelpad=0, color = 'purple')
ax.set_ylabel("y", fontsize=9, labelpad=0, color = 'purple')
ax.xaxis.set_label_coords(0.5, -0.08)
ax.yaxis.set_label_coords(-0.08, 0.5)

# 7. Remove overlapping '0' on y-axis
ax.set_yticks([tick for tick in ax.get_yticks() if tick != 0])

# 8. Customize plot
plt.grid(True)
plt.legend(loc='upper center', bbox_to_anchor=(1.0, 1.15), fontsize=9)
plt.title("Graph of y = 2¬∑sin(2x)", fontsize=20, pad=15, color = 'orange' )
plt.ylim(-3, 3)

# 9. Display
plt.tight_layout()
plt.show()
````

<h3 style = color:#060270;>Graph Gaussian Curve(2)</h3>

````python
#step 0(setup)
import numpy as np
import matplotlib.pyplot as plt
#step 1(defining our function)
x = np.linspace(-10,10,1000)
y = np.exp(-x**2)
#step 3(set up plot)
plt.figure(figsize = (12,6))
plt.style.use('dark_background')
plt.fill_between(x,y,color = 'royalblue', alpha = 0.35)
plt.plot(x, y, label = r"$f(x) = e^{-x^2}$",color = 'b', linewidth = 2)
#step 4(origin centered coordinate axis)
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
#step 5(labels)
ax.set_xlabel("x-axis", fontsize = 15, labelpad = 0, color = "orange")
ax.set_ylabel("y-axis", fontsize = 15, labelpad = 0, color = "orange")
ax.xaxis.set_label_coords(0.55,0.01)
ax.yaxis.set_label_coords(0.49,0.14)
#step 6(removing overlapping zero on y axis)
ax.set_yticks([tick for tick in ax.get_yticks() if tick != 0])
#step 7 (axis arrows)
ax.plot(1,0,">y",transform = ax.get_yaxis_transform(), clip_on = False)
ax.plot(0,1,"^y",transform = ax.get_xaxis_transform(), clip_on = False)
#step 8(customize plot!!)
plt.axhline(0, color = 'gray', linewidth = 1)
plt.axvline(0, color = 'gray', linewidth = 1)
plt.grid(True)
plt.legend(loc='upper center', bbox_to_anchor = (1.0,1.15),fontsize = 15)
plt.title(r"Graph"+" of "+r"$e^{-x^2}$", fontsize = 27, pad = 15, color = 'skyblue')
#step 9(display)
plt.tight_layout()
plt.show()
````

<h4 style = color:orange;font-weight:bold;>Plotting 2 Graphsüåä+üåä</h4>

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4*np.pi ,4*np.pi, 1000)
y = 2*np.sin(2*x)
dy = 4*np.cos(2*x)
plt.plot(x, y, label = 'y = 2sin(2x)', color = 'g', linewidth = 1)
plt.plot(x, dy, label = 'dy = 4cos(2x)', color ='orange', linewidth = 1)
plt.axhline(0, color = 'gray', linewidth = 1)
plt.axvline(0, color = 'gray', linewidth = 1)
plt.title("Function vs. Derivative")
plt.xlabel("x")
plt.ylabel("y and f'(y)")
plt.grid(True)
plt.legend(loc='upper center', bbox_to_anchor=(1.0, 1.15))
plt.ylim(-5, 5)
ax = plt.gca()
ax.spines['bottom'].set_linewidth(1.2)
ax.spines['left'].set_linewidth(1.2)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel("x-axis",fontsize = 7, labelpad = 0)
ax.set_ylabel("y and dy",fontsize = 7, labelpad = 0)
ax.xaxis.set_label_coords(0.57, 0.47)  # x-label: right side, slightly below x-axis
ax.yaxis.set_label_coords(0.48, 0.57)  # y-label: top side, slightly left of y-axis
ax.set_yticks([tick for tick in ax.get_yticks() if tick != 0])
# Get the current Axes
ax = plt.gca()

# Add arrowhead to x-axis (right direction)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)

# Add arrowhead to y-axis (upward direction)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
plt.show()
````

<h3 style = color:purple;>‚ö°Let's plot a piece wise defined function, using the acquired knowledge‚ö°</h3>
 


````python
#step 0: importing
import numpy as np
import matplotlib.pyplot as plt
#step 1: defining domain 
x1 = np.linspace(0, 2, 500)
x2 = np.linspace(2, 4, 200)
x3 = np.linspace(4, 6, 700)
x4 = np.linspace(6, 8, 200)
#step 2: defining our piece wise function
y1 = 2*np.sin(5*x1)
y2 = np.exp(-x2)
y3 = np.cos(10*x3)
y4 = x4 - 8
#step 3: plotting and labelling our function with customizationüé®üñåÔ∏è!
# try to explore extreme cases like linewidth = 0.1, ls = 10.
plt.figure(figsize=(10, 4))# wider graph, for our purpose
plt.plot(x1, y1, label = "y = 2sinx", color="blue", linewidth=3, linestyle = '--')
plt.plot(x2, y2, label = r"$y =e^{-x}$", color="orange", linewidth=4, linestyle = ':')
plt.plot(x3, y3, label = "y = cosx", color="red", linewidth=0.5, linestyle = '-')
plt.plot(x4, y4, label = "y = x - 8", color="purple", linewidth=2, linestyle = '-.')
#üëâeven if you skip remaining steps, these two are mandatory for plotting our function and showing label
plt.legend()
plt.show()
````
