```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4*np.pi, 4*np.pi, 1000)
y = 2*np.sin(2*x)
plt.plot(x, y, label = 'y = 2*np.sin(2*x)', color = 'g', linestyle = '-', linewidth = 2)
plt.plot(0.25*np.pi ,2 ,'ro')
plt.text(0.25*np.pi ,2.1,"It's a local maxima", ha = 'center', fontsize = 9)
plt.axhline(0, color = 'gray', lw = 1)
plt.grid(True)
plt.legend()
plt.title("A 2D Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.ylim(-3, 3)
plt.legend(loc = 'upper center', bbox_to_anchor = (1.0, 1.15))
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4*np.pi ,4*np.pi, 1000)
y = 2*np.sin(2*x)
dy = 4*np.cos(2*x)
plt.plot(x, y, label = 'y = 2sin(2x)', color = 'g', linewidth = 1)
plt.plot(x, dy, label = 'dy = 4cos(2x)', color ='orange', linewidth = 1)
plt.axhline(0, color = 'gray', linewidth = 1)
plt.axvline(0, color = 'gray', linewidth = 1)
plt.title("Function vs. Derivative")
plt.xlabel("x")
plt.ylabel("y and f'(y)")
plt.grid(True)
plt.legend(loc='upper center', bbox_to_anchor=(1.0, 1.15))
plt.ylim(-5, 5)
ax = plt.gca()
ax.spines['bottom'].set_linewidth(1.2)
ax.spines['left'].set_linewidth(1.2)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel("x-axis",fontsize = 7, labelpad = 0)
ax.set_ylabel("y and dy",fontsize = 7, labelpad = 0)
ax.xaxis.set_label_coords(0.57, 0.47)  # x-label: right side, slightly below x-axis
ax.yaxis.set_label_coords(0.48, 0.57)  # y-label: top side, slightly left of y-axis
ax.set_yticks([tick for tick in ax.get_yticks() if tick != 0])
# Get the current Axes
ax = plt.gca()

# Add arrowhead to x-axis (right direction)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)

# Add arrowhead to y-axis (upward direction)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define domain
x = np.linspace(-4 * np.pi, 4 * np.pi, 800)

# 2. Define function(s)
y = 2 * np.sin(2 * x)
# Optional: derivative or second function
# dy = 4 * np.cos(2 * x)

# 3. Set up plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='y = 2sin(2x)', color='green', linewidth=2)
# Optional: add derivative
# plt.plot(x, dy, '--', label="f'(x) = 4cos(2x)", color='orange')

# 4. Origin-centered coordinate axes
ax = plt.gca()
ax.spines['bottom'].set_position('zero')   # x-axis at y=0
ax.spines['left'].set_position('zero')     # y-axis at x=0
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

# 5. Optional: axis arrows
ax.plot(1, 0, ">r", transform=ax.get_yaxis_transform(), clip_on=False)  # x-arrow
ax.plot(0, 1, "^r", transform=ax.get_xaxis_transform(), clip_on=False)  # y-arrow

# 6. Labels (repositioned & resized)
ax.set_xlabel("x", fontsize=9, labelpad=0, color = 'purple')
ax.set_ylabel("y", fontsize=9, labelpad=0, color = 'purple')
ax.xaxis.set_label_coords(0.5, -0.08)
ax.yaxis.set_label_coords(-0.08, 0.5)

# 7. Remove overlapping '0' on y-axis
ax.set_yticks([tick for tick in ax.get_yticks() if tick != 0])

# 8. Customize plot
plt.grid(True)
plt.legend(loc='upper center', bbox_to_anchor=(1.0, 1.15), fontsize=9)
plt.title("Graph of y = 2sin(2x)", fontsize=20, pad=15, color = 'orange' )
plt.ylim(-3, 3)

# 9. Display
plt.tight_layout()
plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Interactive plotting function
def plot_wave(amplitude=2.0, frequency=2.0, phase=0.0):
    x = np.linspace(-4 * np.pi, 4 * np.pi, 800)
    y = amplitude * np.sin(frequency * x + phase)

    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label=f"y = {amplitude}sin({frequency}x + {phase})", color='purple')
    
    plt.axhline(0, color='gray', linewidth=1)
    plt.axvline(0, color='gray', linewidth=1)
    
    # Coordinate-style axes
    ax = plt.gca()
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1))
    plt.grid(True)
    plt.ylim(-5, 5)
    plt.title("Interactive Sine Wave")
    plt.show()

# Make sliders for amplitude, frequency, and phase
interact(plot_wave,
         amplitude=FloatSlider(min=0.5, max=5, step=0.1, value=2.0),
         frequency=FloatSlider(min=0.5, max=5, step=0.1, value=2.0),
         phase=FloatSlider(min=-np.pi, max=np.pi, step=0.1, value=0.0))
```

```python
import matplotlib.pyplot as plt

def hailstone(n):
    seq = [n]
    while n != 1:
        n = n // 2 if n % 2 == 0 else 3 * n + 1
        seq.append(n)
    return seq

# Choose a starting number
start = 27
sequence = hailstone(start)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(range(len(sequence)), sequence, marker='o', linestyle='-', color='purple')
plt.title(f"Hailstone Sequence Starting at {start}")
plt.xlabel("Step")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def sine_plot(A = 1.0, f = 1.0, phase = 0.0):
    x = np.linspace(-4*np.pi, 4*np.pi, 800)
    y = A*np.sin(f*x + phase)
    
    plt.figure(figsize = (8,4))
    plt.plot(x,y,label=f"y = {A:.1f}sin({f:.1f}x + {phase:.1f})", color = 'royalblue')
    plt.axhline(0, color = 'gray', linewidth = 1)
    plt.axvline(0,color = 'gray', linewidth = 1)
    
    ax = plt.gca()
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.title(" Interactive Sine Wave")
    plt.ylim(-5, 5)
    plt.grid(True)
    plt.legend()
    plt.show()         

interact(
    sine_plot,
    A = FloatSlider(min = 0.5, max =5, step = 0.1,value = 2.0),
    f=FloatSlider(min=0.5, max=5, step=0.1, value=2.0),
    phase=FloatSlider(min=-np.pi, max=np.pi, step=0.1, value=0.0)
)

```

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

# Apply dark theme
plt.style.use('dark_background')

# Define the domain and function
x = np.linspace(-8, 8, 800)
A = 1.0
a = 0.5
y = A * np.exp(-a * x**2)

# Create figure and plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, label=rf"$y = {A:.1f}e^{{-{a:.1f}x^2}}$", linewidth=2, color='deepskyblue')
plt.fill_between(x, y, color='deepskyblue', alpha=0.35)

# Axis styling
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1.2)
ax.spines['left'].set_linewidth(1.2)

# Axis arrows
ax.plot(1.01, 0, ">w", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1.01, "^w", transform=ax.get_xaxis_transform(), clip_on=False)

# Labels and grid
ax.set_xlabel("x-axis", fontsize=10, color='lightgray', labelpad=0)
ax.set_ylabel("y-axis", fontsize=10, color='lightgray', labelpad=0)
ax.xaxis.set_label_coords(0.5, -0.08)
ax.yaxis.set_label_coords(-0.06, 0.5)
ax.set_yticks([tick for tick in ax.get_yticks() if tick != 0])

plt.grid(True, color='gray', alpha=0.3)
plt.legend(loc='upper right', fontsize=10)
plt.title(" Elegant Gaussian Curve", fontsize=16, color='violet', pad=10)
plt.tight_layout()
plt.show()
```

```python
#import numpy as np
#import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Domain
x = np.linspace(-6, 6, 1000)

# Setup
fig, ax = plt.subplots(figsize=(10, 5))
line, = ax.plot([], [], lw=2, color='deepskyblue')
fill = ax.fill_between(x, np.zeros_like(x), np.zeros_like(x), color='deepskyblue', alpha=0.3)

ax.set_xlim(-6, 6)
ax.set_ylim(0, 1.2)
ax.set_title("Animated Gaussian with Shaded Area")
ax.grid(True)

# For updating the fill, we keep a reference
def init():
    line.set_data([], [])
    return line, fill

def animate(i):
    a = 0.5 + 0.3 * np.sin(i * 0.1)
    y = np.exp(-a * x**2)
    line.set_data(x, y)

    # Clear and redraw filled area (replace previous one)
    ax.collections.clear()  # remove old fill
    ax.fill_between(x, 0, y, color='deepskyblue', alpha=0.3)

    return line,
    
ani = FuncAnimation(fig, animate, init_func=init, frames=150, interval=50, blit=False)
plt.show()
```
